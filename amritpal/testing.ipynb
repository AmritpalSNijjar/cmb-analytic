```python
planck = analytic_CMB(0.06*0.25, 0.93*0.25, 50, 1.15, 1, 0.054)
```

```python
a_s = planck.a_s


for k in [0.001, 0.02, 0.05, 0.1]:#np.linspace(0.001, 0.1, 10):

    phi_plt = np.array([planck.Phi_ak(a, k) for a in a_s])
    #phi_plt = phi_plt/phi_plt[0]

    plt.semilogx(planck.a_s, phi_plt*(k)**(3/2), label = f"k={round(k, 3)}")

plt.ylim(0, 1.2)
plt.legend()
#plt.xlim(1e-5, 1e-2)
```
